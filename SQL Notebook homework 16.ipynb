```
SELECT
    COUNT(*) AS "All products"
FROM
    FACT_WEB_TRAFFIC wt
    INNER JOIN DIM_PRODUCT p ON wt.PRODUCT_KEY = p.PRODUCT_KEY;
```

SELECT
    category,
    AVG(unit_price) AS average_price


```
SELECT
    category,
    AVG(unit_price) AS average_price
FROM
    DEMO_DW.SALES.DIM_PRODUCT
GROUP BY
    category
ORDER BY
    average_price DESC;
```

SELECT
    segment,
    COUNT(customer_id) AS customer_count


```
SELECT
    segment,
    COUNT(customer_id) AS customer_count
FROM
    DEMO_DW.SALES.DIM_CUSTOMER
GROUP BY
    segment
ORDER BY
    customer_count DESC;
```

SELECT
    category,
    SUM(unit_price) AS total_price


```
SELECT
    category,
    SUM(unit_price) AS total_price
FROM
    DEMO_DW.SALES.DIM_PRODUCT
WHERE
    active = TRUE
GROUP BY
    category
ORDER BY
    total_price DESC;
```

```
SELECT
    category,
    AVG(unit_price) AS avg_price
FROM
    DEMO_DW.SALES.DIM_PRODUCT
GROUP BY
    category
ORDER BY
    avg_price DESC
LIMIT
    2;
```
